```python
#FARS dataset
#I had to locally download and edit .csv names so they wouldn't overwrite
url15 = 'https://raw.githubusercontent.com/safe-routes/Build-SafeRoutes-DS/master/accident2015.csv'
url16 = 'https://raw.githubusercontent.com/safe-routes/Build-SafeRoutes-DS/master/accident2016.csv'
url17 = 'https://raw.githubusercontent.com/safe-routes/Build-SafeRoutes-DS/master/accident2017.csv'
ac15 = pd.read_csv(url15)
ac16 = pd.read_csv(url16) 
ac17 = pd.read_csv(url17)
```

```python
# Dropping observations with invalid LONGITUDE
# Longitudes cannot be less than -180 NOR greater than 180

ac = ac[ac['LONGITUD'] < 180]
```

```python
ac.columns
```

```python
cal = USFederalHolidayCalendar()
holidays = cal.holidays('2015', '2016', '2017')
ac = ac.join(pd.Series(1, index=holidays, name='holiday'))
ac['holiday'].fillna(0, inplace=True)
```

```python
# California accidents coordinates (2015-2017)
county_list = [1, 41, 75]
ac_coord = ac[(ac['COUNTY']==1) | (ac['COUNTY']==41) | (ac['COUNTY']==75) | (ac['COUNTY']==85)| (ac['COUNTY']==81) ]
```
